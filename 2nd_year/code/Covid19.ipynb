```python
import csv
import matplotlib.pyplot as plt


f = open('covid.csv', encoding='cp949')
data = csv.reader(f)
covid = []
date = []

next(data)
next(data)

name = str(input(".  :"))
tt = name.split(".")
month = tt[1]

for row in data:
    if name in row[0]:
        if month == row[0].split(".")[1]:
            covid.append(int(row[1]))
            date.append(row[0].split(".")[2])

plt.figure(figsize=(16,8))
plt.title(f'{tt[0]}-{month if len(month) == 2 else "0"+month }  covid infection person')

plt.plot(date,covid, label = 'covid infection person')
plt.legend()


plt.show()
```
