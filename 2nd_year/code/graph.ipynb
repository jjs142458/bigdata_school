```python
import matplotlib.pyplot as plt
import numpy as np
import random

res = []

for i in range(13):
    res.append(random.randint(1,1000))
print(sorted(res))

res = np.array(res)
print("1/4 "+ str(np.percentile(res,25)))
print("2/4 "+ str(np.percentile(res,50)))
print("3/4 "+ str(np.percentile(res,75)))

plt.boxplot(res)
plt.show()
```

```python
import csv 
import matplotlib.pyplot as plt

f = open('seoultemp.csv', encoding = 'cp949')
data = csv.reader(f)
next(data)

month = [[],[],[],[],[],[],[],[],[],[],[],[]]

for row in data:
    if row[-1] != '':
        month[int(row[0].split('-')[1])-1].append(float(row[-1]))


plt.boxplot(month)

plt.show()
```

```python
import csv 
import matplotlib.pyplot as plt

f = open('seoultemp.csv', encoding = 'cp949')
data = csv.reader(f)
next(data)
day = []

for i in range(31) :
    day.append([])
    
for row in data:
    if row[-1] != '':
        if row[0].split('-')[1] == '08':
            day[int(row[0].split('-')[2])-1].append(float(row[-1]))
            
plt.figure(figsize = (10,5),dpi = 300)
plt.style.use('ggplot')
plt.boxplot(day)
plt.show()
```

```python
import csv

f = open('gender.csv')
data = csv.reader(f)

m = []
f = []
name = str(input("     : "))

for row in data:
    if  name in row[0]:
        for i in range(0, 101):
            m.append(-int(row[i + 3]))
            f.append(int(row[-(i+1)]))
        break
            
f.reverse()
m.reverse()

import matplotlib.pyplot as plt
plt.barh(range(101),f)
plt.barh(range(101),m)
plt.show()
```

```python
import csv

f = open('type.csv')
data = csv.reader(f)
next(data)
t = []
r = []

for row in data:
    t.append(row[2])

r.append(t.count("A"))
r.append(t.count("B"))
r.append(t.count("O"))
r.append(t.count("AB"))

import matplotlib.pyplot as plt
label = ['A','B','O','AB']
plt.axis('equal')
plt.pie(r, labels = label, autopct = '%.1f%%', explode=(0,0,0.1,0))
plt.legend()
plt.show()

# autopct  auto percent 
#          
#        autopct %.1f%% 
#   explode 0  
#     matplotlib api 
```
